```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.preprocessing import StandardScaler
```

```python
# Example dataset (you can modify)
data = {
    "Math": [80, 60, 75, 90, 50],
    "Science": [70, 85, 65, 95, 55],
    "English": [90, 75, 70, 85, 60]
}

df = pd.DataFrame(data)
print("Dataset:\n", df)
```

🔹 Part A: Vector & Matrix Fundamentals

1️⃣ Represent vectors

```python
A = df.iloc[0].values  # Student 1
B = df.iloc[1].values  # Student 2

print("Vector A:", A)
print("Vector B:", B)
```

2️⃣ Norm

```python
norm_A = np.linalg.norm(A)
norm_B = np.linalg.norm(B)

print("Norm A:", norm_A)
print("Norm B:", norm_B)
```

3️⃣ Dot Product

```python
dot_product = np.dot(A, B)
print("Dot Product:", dot_product)
```

4️⃣ Angle Between Vectors

```python
angle = np.arccos(dot_product / (norm_A * norm_B))
angle_deg = np.degrees(angle)

print("Angle (degrees):", angle_deg)
```

5️⃣ Cross Product (3D)

```python
cross_product = np.cross(A, B)
print("Cross Product:", cross_product)
```

6️⃣ Projection of A onto B

```python
projection = (np.dot(A, B) / np.dot(B, B)) * B
print("Projection of A onto B:", projection)
```

🔹 Part B: Matrix Operations

7️⃣ Matrix Formation

```python
M = df.values
print("Matrix M:\n", M)
```

8️⃣ Matrix Addition

```python
print("M + M:\n", M + M)
```

9️⃣ Matrix Multiplication

```python
print("M x M^T:\n", np.dot(M, M.T))
```

🔟 Transpose

```python
print("Transpose:\n", M.T)
```

1️⃣1️⃣ Determinant (square matrix required)

```python
square_matrix = np.dot(M, M.T)
det = np.linalg.det(square_matrix)

print("Determinant:", det)
```

🔹 Part C: Linear Transformations

1️⃣2️⃣ Simple Transformation (Scaling)

```python
scaled = 2 * M
print("Scaled Matrix:\n", scaled)
```

1️⃣3️⃣ Rotation (2D Example using first 2 columns)

```python
theta = np.radians(45)

rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])

rotated = np.dot(df[["Math", "Science"]], rotation_matrix)
print("Rotated Data:\n", rotated)
```

🔹 Part D: Eigenvalues & Decomposition

1️⃣4️⃣ Covariance Matrix

```python
cov_matrix = np.cov(M.T)
print("Covariance Matrix:\n", cov_matrix)
```

1️⃣5️⃣ Eigenvalues & Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

1️⃣6️⃣ LU Decomposition

```python
from scipy.linalg import lu

P, L, U = lu(square_matrix)

print("L:\n", L)
print("U:\n", U)
```

1️⃣7️⃣ SVD

```python
U, S, Vt = np.linalg.svd(M)

print("U:\n", U)
print("Singular Values:\n", S)
print("Vt:\n", Vt)
```

🔹 Part E: Dimensionality Reduction

1️⃣8️⃣ PCA

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(M)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

print("PCA Result:\n", pca_result)
```

1️⃣9️⃣ LDA (Classification)

```python
# Create labels (Above Avg = 1, Below Avg = 0)
df["Average"] = df.mean(axis=1)
df["Label"] = (df["Average"] > df["Average"].mean()).astype(int)

X = df[["Math", "Science", "English"]]
y = df["Label"]

lda = LDA(n_components=1)
lda_result = lda.fit_transform(X, y)

print("LDA Result:\n", lda_result)
```
